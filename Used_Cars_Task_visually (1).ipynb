# Q4
* ## Map the Car Sale Status column (Sold = 1 / UnSold = 0)
* ## Map the Customer Feedback column

```python
df["Car Sale Status"].replace({"Sold" : 1, "Un Sold" : 0}, inplace=True)
```

```python
ax=df["Car Sale Status"].value_counts().plot(kind='bar', color=sns.color_palette("pastel"),figsize=(10,5))
plt.title("Car Sale Status")
plt.ylabel("Count")
plt.xlabel("Status")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
ax=df["Feedback"].value_counts().sort_values(ascending=False).plot(kind="bar", color=sns.color_palette("pastel"),figsize=(10,5))
plt.title("Customer Feedback")
plt.ylabel("Count")
plt.xlabel("Feedback")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
df["Feedback"].replace({"Excellent" : 4, "Good" : 3, "Average" : 2, "Poor" : 1}, inplace=True)
```
