```python
def age_group(age):
  if age>=18 and age<=25:
    return '18-25'
  elif age>=26 and age<=35:
    return '26-35'
  elif age>=36 and age<=45:
      return '36-45'
  elif age>=46 and age<=55:
      return '46-55'
  else:
    return '56+'
df['Age_Group']=df['Age'].apply(age_group)


```

```python
def Previous_Purchase_group(Previous_Purchase):
  if Previous_Purchase>=1 and Previous_Purchase<=10:
    return '1-10'
  elif Previous_Purchase>=11 and Previous_Purchase<=20:
    return '11-20'
  elif Previous_Purchase>=21 and Previous_Purchase<=30:
      return '21-30'
  elif Previous_Purchase>=31 and Previous_Purchase<=40:
      return '31-40'
  else:
    return '41-50'
df['Previous_Purchase_Group']=df['Age'].apply(Previous_Purchase_group)


```

```python
ax=df.pivot_table(index='Gender',columns='Previous_Purchase_Group',values='Previous Purchases',aggfunc=['count']).plot(kind='bar',color=sns.color_palette('pastel'))
for i in ax.containers:
  ax.bar_label(i)
plt.title('Gender vs Previous Purchases vs AVG purchase')
plt.xlabel('Gender')
plt.ylabel('count')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```
