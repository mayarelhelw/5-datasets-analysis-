```python
ax=df["Sales"].agg(["min", "max", "sum", "mean"]).plot(kind="bar", color="red", title="Sales", figsize=(10,5))
for i in ax.containers:
    ax.bar_label(i)
plt.xlabel("Sales")
plt.ylabel("Sales")
plt.show()
```

```python
Sub_Category = df.pivot_table(index="Sub-Category", values="Sales", aggfunc=["mean", "min", "max"])
Sub_Category
```

```python
plt.figure(figsize=(10,5))
ax=df.pivot_table(index="Sub-Category",columns='Category', values="Sales", aggfunc=["mean"]).plot(kind="bar", color=sns.color_palette("Set2"), title="Sub-Category", figsize=(10,5))
plt.xlabel("Sub-Category")
plt.ylabel("Sales")
plt.xticks(rotation=90)
plt.show()

```

```python

office_df = df[df['Category'] == 'Office Supplies']

plt.figure(figsize=(8,5))
sns.scatterplot(data=office_df, x='Discount', y='Sales', alpha=0.5)
plt.title('Discount vs Sales (Office Supplies)')
plt.xlabel('Discount')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()

correlation = office_df['Discount'].corr(office_df['Sales'])
print(f"Correlation between Discount and Sales: {correlation:.2f}")
```

```python
df['Month'] = df['Order Date'].dt.month
office_df = df[df['Category'] == 'Office Supplies']

monthly_sales = office_df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(8,5))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Monthly Sales for Office Supplies')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
office_df = df[df['Category'] == 'Office Supplies']
segment_orders = office_df['Segment'].value_counts()
plt.figure(figsize=(6,4))
segment_orders.plot(kind='bar', color='orange')
plt.title('Order Count by Segment (Office Supplies)')
plt.xlabel('Segment')
plt.ylabel('Number of Orders')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values()
region_sales.plot(kind='barh', color='purple', figsize=(8,5), title='Sales by Region')
plt.xlabel('Sales')
plt.tight_layout()
plt.show()
```

```python
df['Year'] = df['Order Date'].dt.year

yearly_sales = df.groupby('Year')['Sales'].sum()

plt.figure(figsize=(8,5))
yearly_sales.plot(kind='line', marker='o', color='darkblue')
plt.title('Total Sales per Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```
